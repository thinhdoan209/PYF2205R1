```python
from tkinter import *
from tkinter import filedialog

import pyperclip #Dùng clipboard: pip3 install pyperclip
from googletrans import Translator #pip3 install googletrans

sc = Tk()
sc.title("Dịch")

translator = Translator()

#----------------------FUNC---------------------
def browseFiles():
    filename = filedialog.askopenfilename(initialdir = "/",
                                          title = "Chọn file Txt",
                                          filetypes = [("Text files","*.txt")])
    lb4_mh2.config(text = f"Đã chọn file: {filename}")
    fl_L.set(filename)
    
def chuyendoisangFile():
    manhinh1.grid_forget()
    manhinh2.grid(column = 1, row = 1)
    
def chuyendoisangNhap():
    manhinh2.grid_forget()
    manhinh1.grid(column = 1, row = 1)
    
def dichMh1():
    text = t1_mh1.get("1.0", "end-1c")
    text = dichAV(text)
    t2_mh1.config(state = "normal")
    t2_mh1.delete("1.0", END)
    t2_mh1.insert(END, text)
    t2_mh1.config(state = "disable")
    bt3_mh1.grid(column = 1, row = 4, sticky = "e")
    
def dichMh2():
    file = fl_L.get()
    f = open(file, "r")
    text = f.read()
    text = dichAV(text)
    t2_mh2.config(state = "normal")
    t2_mh2.delete("1.0", END)
    t2_mh2.insert(END, text)
    t2_mh2.config(state = "disable")
    bt4_mh2.grid(column = 1, row =5, sticky = "e")
    
def dichAV(text):
    tl = translator.translate(text, src='en', dest='vi')
    return tl.text

def copyMh1():
    pyperclip.copy(t2_mh1.get("1.0", "end-1c"))

def copyMh2():
    pyperclip.copy(t2_mh2.get("1.0", "end-1c"))

#----------------------ENDFUNC---------------------

manhinh1 = Frame(sc)
manhinh1.grid(column = 1, row = 1)

lb1_mh1 = Label(manhinh1, text = "Dịch tiếng Anh qua tiếng Việt.")
lb2_mh1 = Label(manhinh1, text = "ENGLISH")
lb3_mh1 = Label(manhinh1, text = "TIẾNG VIỆT")

t1_mh1 = Text(manhinh1, height = 3)
t2_mh1 = Text(manhinh1, height = 3, state = "disable")

bt1_mh1 = Button(manhinh1, text = "DỊCH", command = dichMh1)
bt2_mh1 = Button(manhinh1, text = "Dịch từ file TXT", command = chuyendoisangFile)
bt3_mh1 = Button(manhinh1, text = "Copy", command = copyMh1)

lb1_mh1.grid(column = 1, row = 1, padx = (10, 10))
lb2_mh1.grid(column = 1, row = 2, sticky = "w")
lb3_mh1.grid(column = 1, row = 4, sticky = "w")

t1_mh1.grid(column = 1, row = 3)
t2_mh1.grid(column = 1, row = 5)

bt1_mh1.grid(column = 1, row = 6, sticky = "e")
bt2_mh1.grid(column = 1, row = 2, sticky = "e")

manhinh2 = Frame(sc)
#manhinh2.grid(column = 1, row =2)
fl_L = StringVar()

lb1_mh2 = Label(manhinh2, text = "Dịch tiếng Anh qua tiếng Việt.")
lb2_mh2 = Label(manhinh2, text = "ENGLISH")
lb3_mh2 = Label(manhinh2, text = "TIẾNG VIỆT")
lb4_mh2 = Label(manhinh2, text = "", textvariable = fl_L)

t2_mh2 = Text(manhinh2, height = 3, state = "disable")

bt1_mh2 = Button(manhinh2, text = "DỊCH", command = dichMh2)
bt2_mh2 = Button(manhinh2, text = "Chọn file", command = browseFiles)
bt3_mh2 = Button(manhinh2, text = "Nhập để dịch", command = chuyendoisangNhap)
bt4_mh2 = Button(manhinh2, text = "Copy", command = copyMh2)

lb1_mh2.grid(column = 1, padx = (10, 10))
lb2_mh2.grid(column = 1, row = 2, sticky = "w")
lb3_mh2.grid(column = 1, row = 5, sticky = "w")
lb4_mh2.grid(column = 1, row = 3, sticky = "w")

t2_mh2.grid(column = 1, row = 6)

bt1_mh2.grid(column = 1, row = 7, sticky = "e")
bt2_mh2.grid(column = 1, row = 4)
bt3_mh2.grid(column = 1, row = 2, sticky = "e")

sc.mainloop()
```

```python
from tkinter import *
from tkinter import filedialog

import pyperclip #Dùng clipboard: pip3 install pyperclip
from googletrans import Translator #pip3 install googletrans

sc = Tk()
sc.title("Dịch")

translator = Translator()

#----------------------FUNC---------------------
def browseFiles():
    filename = filedialog.askopenfilename(initialdir = "/",
                                          title = "Chọn file Txt",
                                          filetypes = [("Text files","*.txt")])
    lb4_mh2.config(text = f"Đã chọn file: {filename}")
    bt2_mh2.grid_forget()
    bt2_mh2.grid(column = 1, row = 3, padx = (10, 10), sticky = "e")
    f = open(filename, "r")
    text = f.read()
    f.close()
    t1_mh2.delete("1.0", END)
    t1_mh2.insert(END, text)
    t1_mh2.grid(column = 1, row = 4, padx = (10, 10))
    
def chuyendoisangFile():
    manhinh1.grid_forget()
    manhinh2.grid(column = 1, row = 1, padx = (10, 10), pady = (10, 10))
    
def chuyendoisangNhap():
    manhinh2.grid_forget()
    manhinh1.grid(column = 1, row = 1, padx = (10, 10), pady = (10, 10))
    
def dichMh1():
    text = t1_mh1.get("1.0", "end-1c")
    text = dichAV(text)
    t2_mh1.config(state = "normal")
    t2_mh1.delete("1.0", END)
    t2_mh1.insert(END, text)
    t2_mh1.config(state = "disable")
    bt3_mh1.grid(column = 1, row = 4, sticky = "e")
    
def dichMh2():
    text = t1_mh2.get("1.0", "end-1c")
    text = dichAV(text)
    t2_mh2.config(state = "normal")
    t2_mh2.delete("1.0", END)
    t2_mh2.insert(END, text)
    t2_mh2.config(state = "disable")
    bt4_mh2.grid(column = 1, row =5, sticky = "e")
    
def dichAV(text):
    tl = translator.translate(text, src='en', dest='vi')
    return tl.text

def copyMh1():
    pyperclip.copy(t2_mh1.get("1.0", "end-1c"))

def copyMh2():
    pyperclip.copy(t2_mh2.get("1.0", "end-1c"))

#----------------------ENDFUNC---------------------

manhinh1 = Frame(sc)
manhinh1.grid(column = 1, row = 1, padx = (10, 10), pady = (10, 10))

lb1_mh1 = Label(manhinh1, text = "Dịch tiếng Anh sang tiếng Việt".upper(), font = ("Tahoma", 18), fg = "brown")
lb2_mh1 = Label(manhinh1, text = "English:", font = ("Tahoma", 15))
lb3_mh1 = Label(manhinh1, text = "Tiếng Việt:", font = ("Tahoma", 15))

t1_mh1 = Text(manhinh1, height = 3, font = ("Tahoma", 15))
t2_mh1 = Text(manhinh1, height = 3, state = "disable", font = ("Tahoma", 15))

bt1_mh1 = Button(manhinh1, text = "Dịch", command = dichMh1, width = 20, font = ("Tahoma", 15), fg = "blue")
bt2_mh1 = Button(manhinh1, text = "Dịch từ file TXT", command = chuyendoisangFile, font = ("Tahoma", 15))
bt3_mh1 = Button(manhinh1, text = "Copy", command = copyMh1, font = ("Tahoma", 15))

lb1_mh1.grid(column = 1, row = 1, padx = (10, 10))
lb2_mh1.grid(column = 1, row = 2, sticky = "w", padx = (10, 10))
lb3_mh1.grid(column = 1, row = 4, sticky = "w", padx = (10, 10))

t1_mh1.grid(column = 1, row = 3, padx = (10, 10))
t2_mh1.grid(column = 1, row = 5, padx = (10, 10))

bt1_mh1.grid(column = 1, row = 6, sticky = "e", padx = (10, 10))
bt2_mh1.grid(column = 1, row = 2, sticky = "e", padx = (10, 10))

manhinh2 = Frame(sc)
#manhinh2.grid(column = 1, row =2)

lb1_mh2 = Label(manhinh2, text = "Dịch tiếng Anh sang tiếng Việt".upper(), font = ("Tahoma", 18), fg = "brown")
lb2_mh2 = Label(manhinh2, text = "English:", font = ("Tahoma", 15))
lb3_mh2 = Label(manhinh2, text = "Tiếng Việt:", font = ("Tahoma", 15))
lb4_mh2 = Label(manhinh2, text = "", font = ("Tahoma", 15))

t1_mh2 = Text(manhinh2, height = 3, font = ("Tahoma", 15))
t2_mh2 = Text(manhinh2, height = 3, state = "disable", font = ("Tahoma", 15))

bt1_mh2 = Button(manhinh2, text = "Dịch", command = dichMh2, width = 20, font = ("Tahoma", 15), fg = "blue")
bt2_mh2 = Button(manhinh2, text = "Chọn File", command = browseFiles, font = ("Tahoma", 15))
bt3_mh2 = Button(manhinh2, text = "Nhập để dịch", command = chuyendoisangNhap, font = ("Tahoma", 15))
bt4_mh2 = Button(manhinh2, text = "Copy", command = copyMh2, font = ("Tahoma", 15))

lb1_mh2.grid(column = 1, padx = (10, 10))
lb2_mh2.grid(column = 1, row = 2, sticky = "w", padx = (10, 10))
lb3_mh2.grid(column = 1, row = 5, sticky = "w", padx = (10, 10))
lb4_mh2.grid(column = 1, row = 3, sticky = "w", padx = (10, 10))

t2_mh2.grid(column = 1, row = 6, padx = (10, 10))

bt1_mh2.grid(column = 1, row = 7, sticky = "e", padx = (10, 10))
bt2_mh2.grid(column = 1, row = 4, padx = (10, 10))
bt3_mh2.grid(column = 1, row = 2, sticky = "e", padx = (10, 10))

sc.mainloop()
```
