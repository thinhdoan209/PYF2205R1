```python
import turtle as t

t.speed(10)

t.penup()
t.goto(-250, -80)
t.pendown()

t.pencolor("#6086BB")
t.pensize(4)

def hcn(a,b):
    for i in range (2):
        t.forward(a)
        t.left(90)
        t.forward(b)
        t.left(90)

hcn(370, 155)

t.forward(142)
hcn(86, 102)

t.penup()
t.goto(-250, -80+155)
t.pendown()
t.left(45)
t.forward(177)
t.right(90)
t.forward(177)
t.backward(82)
t.left(135)
t.forward(50)
t.left(90)
t.forward(24)
t.left(90)
t.forward(26)
t.backward(26)
t.left(90)
t.forward(24)
t.left(90)
t.penup()
t.forward(20)
t.pendown()
t.forward(52)
t.left(90)
t.penup()
t.forward(15)
t.left(90)
t.forward(17)
t.pendown()
t.forward(28)
t.left(90)

def hinhthang(x,y,a,b,h):
    t.penup()
    t.goto(x,y)
    t.pendown()
    t.fillcolor("#808080")
    t.begin_fill()
    t.forward(a/2)
    t.goto(x+b/2,y+h)
    t.backward(b)
    t.goto(x-a/2,y)
    t.forward(a/2)
    t.end_fill()

t.pensize(2)
    
hinhthang(-313, -170, 30, 20, 66)
hinhthang(293, -161, 60, 36, 130)

def tamgiac(x, y, a, color):
    t.penup()
    t.goto(x,y)
    t.pendown()
    t.fillcolor(color)
    t.begin_fill()
    t.forward(a/2)
    t.left(135)
    t.forward(a/(2**0.5))
    t.left(90)
    t.forward(a/(2**0.5))
    t.left(135)
    t.forward(a/2)
    t.end_fill()

t.pencolor("#7C9048")

tamgiac(-313, -170+66, 120, "#525F30")
tamgiac(-313, -170+66+60, 96, "#7C9048")
tamgiac(-313, -170+66+60+48, 55, "#C6D4A1")

tamgiac(293, -161+130, 236, "#525F30")
tamgiac(293, -161+130+118, 186, "#7C9048")
tamgiac(293, -161+130+118+93, 106, "#C6D4A1")

t.done()
```
