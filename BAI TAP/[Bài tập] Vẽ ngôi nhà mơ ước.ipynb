# BÀI LÀM

```python
import turtle
import math

t = turtle.Turtle()
t.speed(10)

turtle.bgcolor("#8ACEE9")

def hbh(x,y,a,b,c,color): #hinh binh hanh goc = c
    t.penup()
    t.goto(x,y)
    t.pendown()
    
    t.fillcolor(color)
    t.begin_fill()
    
    for i in range (2):
        t.forward(a)
        t.left(c)
        t.forward(b)
        t.left(180-c)
        
    t.end_fill()
    
def tamgiac(x,y,a,color):
    t.penup()
    t.goto(x,y)
    t.pendown()
    
    t.fillcolor(color)
    t.begin_fill()
    
    t.forward(a/2)
    t.left(180-45)
    t.forward(a/math.sqrt(2))
    t.left(90)
    t.forward(a/math.sqrt(2))
    t.left(180-45)
    t.forward(a/2)
    
    t.end_fill()

# VẼ NHÀ

hbh(-300, -200, 200, 220, 90, "#2024FB")
hbh(-300+60, -200, 80, 120, 90, "#93ED93")

t.left(30)
hbh(-100, -200, 120, 220, 60, "yellow")
hbh(-65,-90, 50, 70, 60, "#A42B2E")

hbh(-100, 20, 120, 200/math.sqrt(2), 60+45, "orange")
t.right(30)

tamgiac(-300+100, 20, 200, "#FC28FC")

# VẼ MẶT TRỜI

t.penup()
t.goto(100,300)
t.pendown()

for i in range (8):
    t.goto(100,300)
    if i%2==0:
        t.forward(100)
    else:
        t.forward(100/math.sqrt(2))
    t.right(45)

t.fillcolor("yellow")
t.begin_fill()

t.penup()
t.goto(100,300-50)  
t.pendown()

t.circle(50)

t.end_fill()

# VẼ CÂY

hbh(200,-220, 40, 80, 90, "#A42B2E")

tamgiac(200+20,-220+80, 180, "#93ED93")
tamgiac(200+20,-220+80+180/2, 160, "#93ED93")
tamgiac(200+20,-220+80+180/2+160/2, 140, "#93ED93")

turtle.done()
```
