```python
import turtle
import math

t = turtle
t.speed(10)
#t.pensize(2)

t.bgcolor("#8ACEE9")

# draw house

t.penup()
t.goto(-300,-200)
t.pendown()

### Mặt trước 200, 220

t.fillcolor("#2024FB")
t.begin_fill()

for i in range (2):
    t.forward(200)
    t.left(90)
    t.forward(220)
    t.left(90)

t.end_fill()

### Mặt bên 120, 220, c30
t.forward(200)

t.fillcolor("#FEFD38")
t.begin_fill()

t.left(30)
t.forward(120)
t.left(90-30)
t.forward(220)
t.left(90+30)
t.forward(120)

t.end_fill()

### Mái nhà
t.fillcolor("#FC28FC")
t.begin_fill()

t.right(45+30)
t.forward(200/math.sqrt(2))
t.left(90)
t.forward(200/math.sqrt(2))

t.end_fill()

### Mái nhà bên
t.backward(200/math.sqrt(2))

t.fillcolor("#FDA428")
t.begin_fill()

t.right(15+180)
t.forward(120)
t.right(90-15)
t.forward(200/math.sqrt(2))
t.right(90+15)
t.forward(120)

t.end_fill()

### Cửa chính 120, 80

t.penup()
t.goto(-300+60, -200)
t.pendown()

t.fillcolor("#93ED93")
t.begin_fill()

t.goto(-300+60, -200+120)
t.goto(-300+60+80, -200+120)
t.goto(-300+60+80, -200)

t.end_fill()

### Cửa sổ 50,70

t.penup()
t.goto(-300+200,-200+90)
t.backward(40)
t.pendown()

t.fillcolor("#A42B2E")
t.begin_fill()

for i in range(2):
    t.right(90+30)
    t.forward(70)
    t.right(90-30)
    t.forward(50)
    
t.end_fill()

# MẶT TRỜI

t.penup()
t.goto(100,300)
t.pendown()

t.left(180-30)

for i in range (8):
    t.goto(100,300)
    if i%2==0:
        t.forward(100)
    else:
        t.forward(100/math.sqrt(2))
    t.right(45)

t.fillcolor("yellow")
t.begin_fill()

t.penup()
t.goto(100,300-50)  
t.pendown()

t.circle(50)

t.end_fill()

# TREE

t.penup()
t.goto(200,-220)
t.pendown()

### Gốc 40, 80
t.fillcolor("#A42B2E")
t.begin_fill()

for i in range(2):
    t.forward(40)
    t.left(90)
    t.forward(80)
    t.left(90)

t.end_fill()

### LA
def tamgiac(x,y,a):
    t.penup()
    t.goto(x,y)
    t.pendown()
    
    t.fillcolor("#93ED93")
    t.begin_fill()
    
    t.forward(a/2)
    t.left(180-45)
    t.forward(a/math.sqrt(2))
    t.left(90)
    t.forward(a/math.sqrt(2))
    t.left(180-45)
    t.forward(a/2)
    
    t.end_fill()

tamgiac(200+20,-220+80, 180)
tamgiac(200+20,-220+80+180/2, 160)
tamgiac(200+20,-220+80+180/2+160/2, 140)

t.done()
```
